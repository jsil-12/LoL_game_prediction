```python
import numpy as np
import pandas as pd

def read_league_data(file="2017-summer-match-data-OraclesElixir-2017-08-07.csv",path="C:\\Users\\joey\\__Notebooks__\\python notebooks\\LoL_game_prediction\\"):
    return pd.read_csv(path+file, encoding = "ISO-8859-1")
    
def round_weeks(data):
    try:
        data.loc[:,"week"] = data["week"].astype(np.double).round()
    except ValueError:
        data["week"] = pd.to_numeric(data["week"],errors="coerce").round()
    data = data.dropna(subset=["week"])
    return data
    
def iter_weeks(data,returnDF=False):
    weeks = np.unique(data["week"])
    if returnDF:
        for wk in weeks:
            yield data.loc[data["week"] == wk]
    else:
        for wk in weeks:
            yield wk
        
def iter_games(data):
    games = np.unique(data["gameid"])
    for game in games:
        yield game

#input a gameDF and a teamDF   
def add_players(game,team,start=1):
    for playerid in range(start,start+5):
        t = game.loc[game["playerid"]==playerid,["position","champion"]]
        team.loc[:,t["position"].iloc[0]] = t["champion"].iloc[0]
        team.loc[:,t["position"].iloc[0]+"_champ_strength"] = 0
    return team
            
def summarize_game(league,game):
    team1 = league.loc[(league["gameid"]==game) & (league["playerid"]==100)]
    team2 = league.loc[(league["gameid"]==game) & (league["playerid"]==200)]
    teamnames = sorted([str(team1["team"].iloc[0]), str(team2["team"].iloc[0])])
    versus = teamnames[0] + "~" + teamnames[1]
    print(versus)
    team1.loc[:,"matchup"] = versus
    team2.loc[:,"matchup"] = versus
    team1 = add_players(league.loc[(league["gameid"]==game)],team1,start=1)
    team2 = add_players(league.loc[(league["gameid"]==game)],team2,start=6)
    return team1, team2

def summarize_all_games(league):
    all_games = []
    for game in iter_games(league):
        teams = summarize_game(league,game)
        all_games.append(teams[0])
        all_games.append(teams[1])
    return pd.concat(all_games)

def assign_matches(league):
    matchID=1
    all_matches=[]
    for week in iter_weeks(league,returnDF=True):
        matchups = np.unique(week["matchup"])
        for matchup in matchups:
            match = week.loc[week["matchup"]==matchup]
            match.loc[:,"matchID"] = matchID
            match = set_match_win(match)
            matchID += 1
            all_matches.append(match)
    return pd.concat(all_matches)

def set_match_win(match):
    result=[]
    teams = np.unique(match["team"])
    for team in teams:
        t = match.loc[match["team"]==team]
        score = t.loc[:,"result"].sum()
        if  score == 2:
            t.loc[:,"matchResult"] = 1
        else:
            t.loc[:,"matchResult"] = 0
        result.append(t)
    return pd.concat(result)
                      
def process_league(data, league="NALCS", summarize=True):
    league = data.loc[data["league"] == league]
    #print(league)
    league = round_weeks(league)
    if summarize:
        league = summarize_all_games(league)
        league = assign_matches(league)
        #league = set_match_wins(league)
    return league

    #league = summarize_games(league)
print("...done...")    
```

```python
def summarize_games(data,uniquefeatures=["result","gamelength","fb","fbtime","fdtime","teamdragkills","oppdragkills",
                                         "firedrakes","waterdrakes","earthdrakes","airdrakes","elders","herald","fbaron",
                                         "fbarontime","teambaronkills","oppbaronkills","ft","fttime","firstmidouter","firsttothreetowers",
                                         "teamtowerkills","opptowerkills"],
                    meanfeatures=["dmgshare","earnedgoldshare"],
                    sumfeatures=["k","d","a","doubles","triples","quadras","pentas","wards","wardkills","wcpm","visionwards",
                                 "totalgold","earnedgpm","goldspent","minionkills","monsterkills","monsterkillsownjungle",
                                 "monsterkillsenemyjungle","cspm","goldat10","oppgoldat10","gdat10","goldat15","oppgoldat15",
                                 "gdat15","dmgtochamps","dmgtochampsperminute","wpm"],
                   idfeatures=["team","week","match","game","matchID"], return_all = False):
    team_data = {}
    all_data = []
    datacols = uniquefeatures+meanfeatures+sumfeatures+idfeatures+["opponent"]
    for game in iter_all_games(data):
        teams = np.unique(game["match"])[0].split("-")
        opps = list(reversed(teams))
        for t,o in zip(teams,opps):
            print(t,o)
        #for t in np.unique(game["team"]):
            team = game.loc[(game["team"] == t) & ~(game["player"]=="Team")]
            unique_feats = team[uniquefeatures].apply(pd.to_numeric, axis=0).iloc[0]
            mean_feats = team[meanfeatures].apply(pd.to_numeric, axis=0).std(axis=0)
            sum_feats = team[sumfeatures].apply(pd.to_numeric,args=('coerce',),axis=0).sum()
            id_feats = team[idfeatures].iloc[0]
            
            all_feats = list(unique_feats.values) + list(mean_feats.values) + list(sum_feats.values) + list(id_feats.values) + [o]
            data = pd.DataFrame(all_feats,datacols).T
            #print(data)
            if return_all:
                all_data.append(data)
                
            if t in team_data.keys():
                team_data[t].append(data)
            else:
                team_data[t] = [data]
    for t in team_data:
        team_data[t] = pd.concat(team_data[t],ignore_index=True)
        
    if return_all:
        return pd.concat(all_data,ignore_index=True)
    else:
        return team_data
    
def convert_to_numeric(df):
    return df.apply(pd.to_numeric,args=('ignore',))
    
def numeric_columns(df,not_numeric=False,exceptions=[]):
    df = convert_to_numeric(df)
    df = df.drop(exceptions, axis=1)
    #print(df.dtypes)
    if not_numeric:
        result = list(df.select_dtypes(exclude=[np.number]).columns.values)
    else:
        result = list(df.select_dtypes(include=[np.number]).columns.values)
    return result

def summarize_matches_from_games(teamDict,allDF=False):
    teams = {}
    #matchID = 0
    for t in teamDict:
        teamsum = []
        tData = teamDict[t]
        for match in iter_all_matches(tData):
            sumDF = summarize_columns(match, numeric=True, IDfeats=True, sumFeatures=[])
            #sumDF["matchID"] = str(matchID) + "_K17"
            record, gamenums, wins = calc_record(match)
            sumDF["matchwin"] = record
            sumDF["numberofgames"] = gamenums
            sumDF["gamewins"] = wins
            teamsum.append(sumDF)
            #matchID += 1
        teams[t] = pd.concat(teamsum,ignore_index=True)
    return teams

def summarize_columns(data,numeric=True,IDfeats=True,features=[], sumFeatures=[]):
    allcols = []
    datacols = []
    summarydata=[]
    if numeric:
        if features:
            datacols = numeric_columns(data[features])
        else:
            if sumFeatures:
                datacols = numeric_columns(data)
                datacols = list(np.setdiff1d(np.array(datacols),np.array(sumFeatures)))
            else:
                datacols = numeric_columns(data)
            
        allcols += datacols
        summarydata += list(data[datacols].mean().values)
    if IDfeats:
        IDcols = numeric_columns(data,not_numeric=True)
        allcols += IDcols
        ID= data[IDcols].iloc[0]
        summarydata += list(ID.values)
        
    if sumFeatures:
        allcols += sumFeatures
        summarydata += list(data[sumFeatures].sum().values)
    
    re_ = pd.DataFrame(summarydata,allcols).T
    return re_
            
def calc_record(match):
    record = match["result"].mode().values[0]
    gamenums = len(match)
    wins = len(match.loc[match["result"]==1])
    return record, gamenums, wins

def calc_win_percentages(data):
    data["gameWinPercentage"] = np.mean(data["gamewins"])
    data["matchWinPercentage"] = np.mean(data["matchwin"])
    return data

def cummulative_summary(data,first=1,last=-1,
                        summarize=True,
                        numeric=True,
                        IDfeats=True,
                        features=None,
                        sumFeatures=["matchwin","numberofgames","gamewins"]):
    '''
    summarizes the features from start week to end week inclusive
    '''
    weeks = np.unique(data["week"])
    if (last == -1):
        lastwk = weeks[-1]
    else:
        lastwk = weeks[last-1]
    firstwk = weeks[first-1]
    tdata = data.loc[(data["week"]>= firstwk) & (data["week"]<= lastwk)]
    if summarize:
        tdata = summarize_columns(tdata,numeric=numeric, IDfeats=IDfeats, features=features, sumFeatures=sumFeatures)
        tdata = calc_win_percentages(tdata)
        tdata["week"] = str(firstwk)+"-"+str(lastwk)
        #tdata["gamewinperc"] = tdata["gameWins"] / tdata["numberOfGames"]
        
    return tdata

def main_summarize_games(splits,matches=True):
    results = []
    for split in splits:
        gamestats = summarize_games(split)
        #na_team_stats = summarize_games(na_games)
        #na_sp_team_stats = summarize_games(na_sp_games)
        #eu_sp_team_stats = summarize_games(eu_sp_games)
        #eu_team_stats = summarize_games(eu_games)
        results.append(gamestats)
        if matches:
            match = summarize_matches_from_games(gamestats)
            #na_match_dict = summarize_matches_from_games(na_team_stats)
            #na_sp_match_dict = summarize_matches_from_games(na_sp_team_stats)
            #eu_sp_match_dict = summarize_matches_from_games(eu_sp_team_stats)
            #eu_match_dict = summarize_matches_from_games(eu_team_stats)
            results.append(match)
    return results

#nasu, nasp, eusu, eusp = main_summarize_games([na_games, na_sp_games, eu_games, eu_sp_games],matches=False)
print(nasu["Cloud9"]["result"])
print("...summarize data...")
```

```python
def versus(team1,team2,matches_record,matchID):
    #print(team1["matchID"].values[0])
    #team1 = convert_to_numeric(team1)
    #team2 = convert_to_numeric(team2)
    #print(team1.dtypes)
    #print(team2.dtypes)
    a = team1[numeric_columns(team1)]
    b = team2[numeric_columns(team2)]
    c = pdsubtract(a,b)
    
    #cvals = a.values - b.values
    #ccols = a.columns.values
    
    match = matches_record.loc[(matches_record["team"]==team1["team"].values[0]) & (matches_record["matchID"] == matchID)]
    
    c["match"] = match["team"].values[0] + "-" + match["opponent"].values[0] #team1["team"].values[0] + "-" + team2["team"].values[0]
    c["week"] = match["week"].values[0]
    c["matchID"] = matchID
    c["team"] = match["team"].values[0]
    c["opponent"] = match["opponent"].values[0]
    #print(team1["team"])
    c["matchResult"] = get_match_win(matches_record,c["matchID"].values[0],c["team"].values[0])
    
    #vers["matchwin"] = team1["matchwins"]
    return c

def pdsubtract(a,b):
    cvals = a.values - b.values
    ccols = list(a.columns.values)
    return pd.DataFrame(cvals,columns = ccols)

def versus_all_matches(matchData,schedule,matches_record,cummulative="all"):
    versstats = []
    if cummulative == "all":
        print("not implemented")
        
    elif cummulative == "none":
        print("not implemented")
        
    else:
        numOfWeeks = len(np.unique(schedule["week"].values))
        #print(np.unique(schedule["week"].values))
        wkspan = cummulative
        for i in range(cummulative,numOfWeeks):
            stwk = i - wkspan + 1
            endwk = cummulative
            matchwk = endwk + 1
            cummulative += 1
            
            for index, match in schedule.loc[schedule["week"] == matchwk].iterrows():
                team, opp = match["match"].split("-")
                team1 = cummulative_summary(matchData[team],first=stwk,last=endwk,summarize=True,numeric=True,IDfeats=True,features=None,sumFeatures=["matchwin","numberofgames","gamewins"])
                team2 = cummulative_summary(matchData[opp],first=stwk,last=endwk,summarize=True,numeric=True,IDfeats=True,features=None,sumFeatures=["matchwin","numberofgames","gamewins"])
                matchID = match["matchID"]
                vers = versus(team1,team2,matches_record,matchID)
                versstats.append(vers)
                
                #matchData["team"]
                #print(match["match"])
                #print(match["matchID"])
                #print(match["match"])
                #print(vers)
        return pd.concat(versstats,ignore_index=True)

#print(_clg_v_tl)

#print(_clg_v_dig.columns)

def combine_league_splits(leagues):
    return pd.concat(leagues,ignore_index=True)

na_versDF = versus_all_matches(na_match_dict,na_schedule,na_match_record,cummulative=2)
eu_versDF = versus_all_matches(eu_match_dict,eu_schedule,eu_match_record,cummulative=2)

na_sp_versDF = versus_all_matches(na_sp_match_dict,na_sp_schedule,na_sp_match_record,cummulative=2)
eu_sp_versDF = versus_all_matches(eu_sp_match_dict,eu_sp_schedule,eu_sp_match_record,cummulative=2)

versDF = combine_league_splits([na_versDF,eu_versDF,na_sp_versDF,eu_sp_versDF])

print(versDF)
#print(versDF.shape)
versDF_f = versDF.fillna(value=0.0)
#print(versDF_f.dropna().shape)
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

def train_logistic_regression(train_x, train_y):
    """
    Training logistic regression model with train dataset features(train_x) and target(train_y)
    :param train_x:
    :param train_y:
    :return:
    """
    
    logistic_regression_model = LogisticRegression()
    logistic_regression_model.fit(train_x, train_y)
    return logistic_regression_model

def model_accuracy(trained_model, features, targets):
    """
    Get the accuracy score of the model
    :param trained_model:
    :param features:
    :param targets:
    :return:
    """
    accuracy_score = trained_model.score(features, targets)
    return accuracy_score


versDF_features = numeric_columns(versDF)
target = "matchResult"

training_features = versDF_features.remove(target)
#training_features = ["k","d","a","dmgtochamps","fb","fbaron","fbarontime","fbtime","fdtime","ft","fttime","game","gdat10","gdat15"]
#print(training_features)
#print(target)

train_x, test_x, train_y, test_y = train_test_split(versDF_f_std[training_features], versDF_f_std[target], train_size=0.7)

print(train_x)
print(train_y)

print("train_x size %s" %(str(train_x.shape)))
print("train_y size %s" %(str(train_y.shape)))
print("test_x size %s" %(str(test_x.shape)))
print("test_y size %s" %(str(test_y.shape)))

```

```python
from sklearn.decomposition import PCA

def principal_component_analysis(train_X,test_X,nComps=3,measures=["Area","Mean","Circ.","IntDen","Median","Skew","Kurt","Solidity"],scale=False):
    pca=PCA(n_components=nComps)
    pca_train = pca.fit_transform(train_X[measures])
    pca_test = pca.transform(test_X[measures])
    
    print("The principal components explain %s of the variance in the data." %(np.array_str(pca.explained_variance_ratio_)))
    _var = pd.DataFrame(pca.components_.T, index=measures)
    
    return pca_train,pca_test,pca
    
pca_train_x, pca_test_x, model = principal_component_analysis(train_x,test_x,nComps=57,measures=training_features)
#print(pca_train_x)
```
