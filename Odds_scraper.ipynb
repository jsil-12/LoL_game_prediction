```python
import re
from bs4 import BeautifulSoup
import requests

base_url = "http://www.oddsportal.com/esports/world/league-of-legends-championship-series-2017/"

url = base_url

r = requests.get(url)
content = re.sub(r'(?m)^\<!--.*\n?', '', r.content)
content = re.sub(r'(?m)^\-->.*\n?', '', content)
soup = BeautifulSoup(content, 'html.parser')
print (len(soup.find_all('table')))
for table in soup.find_all('table'):
    if table.findParent("table") is None:
        print ("\n\n", str(table))
```

```python
from bs4 import BeautifulSoup, Comment
import requests, re

url = 'http://www.oddsportal.com/esports/world/league-of-legends-championship-series-2017/results/'

def make_soup(url):
    r = requests.get(url)
    soupdata = BeautifulSoup(r.text, 'lxml')
    return soupdata

soup = make_soup(url)
comments = soup.findAll(text=lambda text:isinstance(text, Comment))
print(comments)
rx = re.compile(r'<table.+?id="tournamentTable".+?>[\s\S]+?</table>')

for comment in comments:
    try:
        table = rx.search(comment.string).group(0)
        print(table)
        # break the loop if found
        break
    except:
        print("pass")
        pass
```
